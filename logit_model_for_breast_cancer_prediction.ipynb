```python
#5. Scaling the Numerical features
#Scaling is one of the important thing in the pre-processing as it makes thigs like optimization and calculation easy by scaling the values between a
#certain range like (0,1) or (-1, 1)
#Here we use the range from (0, 1)
```

```python
#Here we are using the MinMaxScaler but we can also use StandardScaler for this step
from sklearn.preprocessing import MinMaxScaler 
```

```python
scaler = MinMaxScaler() #Creating an instance of class MinMaxScaler  
```

```python
""" There are only two sub-steps: 1. Fitting the data to the instance and 2. Transform the values or scale the values """
scaler.fit(raw_df[numeric_cols])
```

```python
print("Minimum: ")
list(scaler.data_min_)
```

```python
#We can now separately scale the training, validation and test sets using the `transform` method of `scaler`.
train_inputs[numeric_cols] = scaler.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scaler.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scaler.transform(test_inputs[numeric_cols])
```

```python
#For visualisation of confusion matrix we can plot by using the helper function.
def predict_and_plot(inputs, targets, name = ''):
    preds = model.predict(inputs) 

    accuracy = accuracy_score(targets, preds)
    print("Accuracy: {:.2f}%".format(accuracy*100))

    cf = confusion_matrix(targets, preds, normalize='true')
    plt.figure()
    sns.heatmap(cf, annot=True)
    plt.xlabel('Predictions') 
    plt.ylabel('Target')
    plt.title('{} Confusion Matrix'.format(name))
    
    return preds
```

```python
train_preds = predict_and_plot(X_train, train_targets, 'Training Set')
```

```python
"""Let's test our model on validation set and see the accuracy on it."""    
val_preds = predict_and_plot(X_val, val_targets, 'Validation Set')
```

```python
test_preds = predict_and_plot(X_test, test_targets, 'Test Set')
```

```python
#Accuracies on Training Set, Validation Set and Test Set    
print("Accuracy on Training Set: ", (accuracy_score(train_targets, train_preds))*100 , "%")
print("Accuracy on Validation Set: ", (accuracy_score(val_targets, val_preds))*100 , "%")
print("Accuracy on Test Set: ", (accuracy_score(test_targets, test_preds))*100 , "%")
```

```python
"""
So our model prediction is around 96 % on test set, therefore it is a nice and accurate model. But it is not end here, as we have to do test
the model on some extra edge cases for determining more accuracy of our model.  
"""
```

```python
""" We have to make two more model called 'random' or 'dumb' models for testing our model"""
def random_guess(inputs):
    return np.random.choice(["B", "M"], len(inputs))
```

```python
#This function is for all benign cases or all non_cancerous cases
def all_B(inputs):
    return np.full(len(inputs), "B")
```

```python
#Accuracy of the random guess
acc = accuracy_score(test_targets, random_guess(X_test))
print("Accuracy of random choice = ", (acc)*100, "%")
```

```python
#Accuracy of all Benign
acc_B = accuracy_score(test_targets, all_B(X_test))
print("Accuracy of all Benign = ", (acc_B)*100, "%")
```

```python
""" As we can See, our model is better than the edge cases with higher accuracy on test set."""
```
